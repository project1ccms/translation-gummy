```python
from deep_translator import GoogleTranslator

sentences = ["ฉันต้องการความรักของคุณ"]

thai = [GoogleTranslator(source='th', target='en').translate(s) for s in sentences]

print(thai)
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "Helsinki-NLP/opus-mt-th-en"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

text = "ฉันต้องการความรักของคุณ"

tokens = tokenizer(text, return_tensors="pt", padding=True)

translated = model.generate(**tokens)

output = tokenizer.decode(translated[0], skip_special_tokens=True)

print(output)
```

```python
import torch
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-th-en")
model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-th-en", dtype=torch.float16, attn_implementation="sdpa", device_map="auto")

inputs = tokenizer("ฉันต้องการความรักของคุณ", return_tensors="pt").to(model.device)
outputs = model.generate(**inputs, cache_implementation="static")
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```

```python
import random
import re


sample_rate = 0.1

with open("OpenSubtitles.en-th.en", encoding="utf-8") as f_en, \
     open("OpenSubtitles.en-th.th", encoding="utf-8") as f_th, \
     open("subset.en", "w", encoding="utf-8") as out_en, \
     open("subset.th", "w", encoding="utf-8") as out_th:

    for en, th in zip(f_en, f_th):
        if random.random() < sample_rate:
            out_en.write(en)
            out_th.write(th)


```

```python
from langdetect import detect, DetectorFactory



import unicodedata
def normalize(text):
    text = unicodedata.normalize("NFKC", text)
    text = text.strip().lower()
    text = re.sub(r"\s+", "", text)
    text = re.sub(r"http\S+", "" , text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r"<.*?>", "", text) 
    text = re.sub(r"(.)\1{5,}", r"\1", text)
         # remove HTML
    return text

def valid_len(en, th):
    en_len = len(en.split())
    th_len = len(th)
    
    if en_len < 2 or th_len < 2:
        return False
    if en_len > 100 or th_len > 300:
        return False
    if th_len / en_len > 5 or en_len / th_len > 5:
        return False
    return True
seen = set()


def detect_lang_safe(text):
    try:
        return detect(text)
    except:
        return "unknown"

def is_valid_lang(en, th):
    return detect_lang_safe(en) == "en" and detect_lang_safe(th) == "th"

en = normalize(en)
th = normalize(th)



if not valid_len(en, th):
    continue
if not is_valid_lang(en, th):
    continue
pair = (en, th)

if pair in seen:
    continue
seen.add(pair)
out_en.write(en + "\n")
out_th.write(th + "\n")

total = 0
kept = 0

kept += 1

# Progress log
if total % 100000 == 0:
    print(f"Processed: {total}, Kept: {kept}")

    print("\n✅ DONE")
    print(f"Total lines processed: {total}")
    print(f"Final clean pairs: {kept}")
```
